```python
import pandas as pd
occupation=pd.read_csv('occupation.csv')
occupation
```

```python
X=occupation.drop(columns=['occupation'])
X
```

```python
y=occupation['occupation']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X,y)
prediction=model.predict([[30,0,2000]])
prediction
```
